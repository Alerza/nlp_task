```python
import pandas as pd
final_voacb = pd.read_csv('/content/drive/MyDrive/sentencepiesese/vocab2llamaEmbedding/vocab16000')
```

```python
import numpy as np
def tokne2fasttext(token):
    ids = model.get_word_id(token)
    if ids>= 0 :
        fasttextID = ids
        fasttextEMBD= model.get_input_vector(ids)
        done = 'direct'
    else:
        ctoken = token.replace('▁','')
        ids = model.get_word_id(ctoken)
        if ids>= 0 :
            fasttextID = ids
            fasttextEMBD= model.get_input_vector(ids)
            done = 'rm_'
        else:
            fasttextID = ids
            fasttextEMBD= np.nan
            done = 'NotAvlbe'

    return fasttextID , fasttextEMBD , done

#sample['tokenizer'] = sample['token'].apply(tokne2tokenize)
final_voacb[['fasttextID','fasttextEMBD', 'done']] = final_voacb['token'].apply(tokne2fasttext).apply(pd.Series)
final_voacb
```
